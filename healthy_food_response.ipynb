```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("response.csv")

print("First 5 rows:\n", df.head())

print("\nColumns:\n", df.columns)
```

```python
print("\nProduct Preference:\n", df['product'].value_counts())
print("\nPrice Preference:\n", df['price'].value_counts())
print("\nCategory Preference:\n",df['category'].value_counts())
print("\nFactor:\n", df['factor'].value_counts())
print("\nBarrier:\n", df['barrier'].value_counts())
print("\nBuy Source:\n", df['source'].value_counts())
```

```python
df['product'].value_counts().plot(kind='bar')
plt.title("Product Preference")
plt.show()

df['price'].value_counts().plot(kind='bar')
plt.title("Price Preference")
plt.show()

df['factor'].value_counts().plot(kind='bar')
plt.title("Key Buying Factor")
plt.show()
```

```python
print("\nProduct vs Price:\n", pd.crosstab(df['product'], df['price']))

print("\nProduct vs Factor:\n", pd.crosstab(df['product'], df['factor']))
```
